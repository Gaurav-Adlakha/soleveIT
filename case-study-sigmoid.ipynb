1. **Class weights**: Add weights to the minority class during model training so that misclassifying the minority class has a higher penalty:

```python
# For many sklearn models

model = RandomForestClassifier(class_weight=class_weights)
```

2. **Evaluation metrics**: Use metrics better suited for imbalanced data:
   - Precision, recall, F1-score instead of accuracy
   - ROC-AUC and PR-AUC (precision-recall area under curve)

3. **Threshold adjustment**: After training, adjust the prediction threshold to optimize for a specific metric:

```python
# Instead of using default 0.5 threshold
y_pred = (model.predict_proba(X_test)[:,1] > 0.3)  # Lower threshold to catch more positive cases
```
