```
def find_matches(ground_truth, ai_suggestions):
    "Find matches between ground truth tables and AI suggestions"
    exact_matches = []
    partial_matches = []
    
    # Convert ground truth to a list of source tables
    gt_tables = ground_truth['source_table'].tolist()
    
    # Check each AI suggestion against ground truth
    for ai_table in ai_suggestions:
        # Check for exact matches
        if ai_table in gt_tables:
            exact_matches.append((ai_table, ai_table, 'exact', 1.0))
        else:
            # Check for partial matches
            for gt_table in gt_tables:
                # Case 1: AI table is substring of ground truth
                if ai_table.lower() in gt_table.lower():
                    score = len(ai_table) / len(gt_table)  # Rough similarity score
                    partial_matches.append((gt_table, ai_table, 'ai_in_gt', score))
                
                # Case 2: Ground truth is substring of AI table
                elif gt_table.lower() in ai_table.lower():
                    score = len(gt_table) / len(ai_table)  # Rough similarity score
                    partial_matches.append((gt_table, ai_table, 'gt_in_ai', score))
                
                # Case 3: Check for word-level similarity
                else:
                    # Split into words and check for common words
                    gt_words = set(gt_table.lower().replace('_', ' ').split())
                    ai_words = set(ai_table.lower().replace('_', ' ').split())
                    common_words = gt_words.intersection(ai_words)
                    
                    if common_words:
                        # Calculate Jaccard similarity
                        score = len(common_words) / len(gt_words.union(ai_words))
                        if score > 0.2:  # Only consider matches with some meaningful similarity
                            partial_matches.append((gt_table, ai_table, 'word_similarity', score))
    
    # Sort partial matches by score
    partial_matches.sort(key=lambda x: x[3], reverse=True)
    
    return exact_matches, partial_matches


def table_matcher(ground_truth, ai_suggestions, min_sim_score=0.2):
    "Match tables between ground truth and AI suggestions with similarity scoring"
    gt_tables = ground_truth['source_table'].tolist()
    exact_matches, partial_matches = [], []
    
    for ai_table in ai_suggestions:
        if ai_table in gt_tables: exact_matches.append((ai_table, ai_table, 'exact', 1.0))
        else:
            for gt_table in gt_tables:
                if ai_table.lower() in gt_table.lower(): 
                    score = len(ai_table)/len(gt_table)
                    if score > min_sim_score: partial_matches.append((gt_table, ai_table, 'ai_in_gt', score))
                elif gt_table.lower() in ai_table.lower(): 
                    score = len(gt_table)/len(ai_table)
                    if score > min_sim_score: partial_matches.append((gt_table, ai_table, 'gt_in_ai', score))
                else:
                    gt_words = set(gt_table.lower().replace('_', ' ').split())
                    ai_words = set(ai_table.lower().replace('_', ' ').split())
                    common = gt_words.intersection(ai_words)
                    if common:
                        score = len(common) / len(gt_words.union(ai_words))
                        if score > min_sim_score: partial_matches.append((gt_table, ai_table, 'word_similarity', score))
    
    partial_matches.sort(key=lambda x: x[3], reverse=True)
    return dict(exact=exact_matches, partial=partial_matches)

matches = table_matcher(gr_tru, ai_match, min_sim_score=0.35)
matches

def format_matches_with_colors(matches):
    "Convert match dictionary to styled pandas DataFrames with color highlighting"
    cols = ['Ground Truth', 'AI Suggestion', 'Match Type', 'Score']
    
    # Format exact matches in red
    if matches['exact']:
        exact_df = pd.DataFrame(matches['exact'], columns=cols)
        exact_df = exact_df.style.set_properties(**{'background-color': 'red', 'color': 'white'})
    else:
        exact_df = pd.DataFrame(columns=cols)
    
    # Format partial matches in yellow
    if matches['partial']:
        partial_df = pd.DataFrame(matches['partial'], columns=cols)
        partial_df = partial_df.style.set_properties(**{'background-color': 'yellow', 'color': 'black'})
    else:
        partial_df = pd.DataFrame(columns=cols)
    
    return exact_df, partial_df

exact_matches, partial_matches = format_matches_with_colors(matches)

print("Exact Matches:")
display(exact_matches)

print("Partial Matches:")
display(partial_matches)
```

```
def match_all_tables(ground_truth_df, ai_suggestions_df, min_sim_score=0.35):
    "Match all tables between ground truth and AI suggestions"
    results = {}
    
    for target_table in ground_truth_df.index.unique():
        if target_table in ai_suggestions_df.index:
            gt_tables = ground_truth_df.loc[target_table, to_keep]
            ai_tables = ai_suggestions_df.loc[target_table]
            
            # Handle case where there's only one row for this target
            if isinstance(gt_tables, pd.Series):
                gt_tables = pd.DataFrame([gt_tables], index=[target_table])
            
            # Convert AI suggestions to list, filtering out NaN values
            if isinstance(ai_tables, pd.Series):
                ai_tables = ai_tables.dropna().tolist()
            
            matches = table_matcher(gt_tables, ai_tables, min_sim_score)
            results[target_table] = matches
    
    return results
```

```
def summarize_matches(match_results):
    "Create a summary of match results"
    summary = {}
    
    for target_table, matches in match_results.items():
        exact_count = len(matches['exact'])
        partial_count = len(matches['partial'])
        summary[target_table] = {
            'exact_matches': exact_count,
            'partial_matches': partial_count,
            'total_matches': exact_count + partial_count
        }
    
    return pd.DataFrame.from_dict(summary, orient='index')
```

```
def display_all_matches(match_results):
    "Display all matches with details"
    all_exact_matches = []
    all_partial_matches = []
    
    for target_table, matches in match_results.items():
        # Add target table info to each match
        for match in matches['exact']:
            all_exact_matches.append((target_table,) + match)
        
        for match in matches['partial']:
            all_partial_matches.append((target_table,) + match)
    
    # Create DataFrames
    cols = ['Target Table', 'Ground Truth', 'AI Suggestion', 'Match Type', 'Score']
    
    exact_df = pd.DataFrame(all_exact_matches, columns=cols)
    partial_df = pd.DataFrame(all_partial_matches, columns=cols)
    
    # Sort by target table and score
    exact_df = exact_df.sort_values(['Target Table', 'Score'], ascending=[True, False])
    partial_df = partial_df.sort_values(['Target Table', 'Score'], ascending=[True, False])
    
    return exact_df, partial_df

exact_matches, partial_matches = display_all_matches(all_matches)

print("Exact Matches:")
display(exact_matches)

print("Partial Matches:")
display(partial_matches)
```
